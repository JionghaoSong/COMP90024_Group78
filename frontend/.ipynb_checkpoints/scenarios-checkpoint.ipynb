## Show liquor shop on map

```python
import json
with open('./liquor_venues.json', 'r') as file:
    data = json.load(file)
    liquor_venues = data['features']

print(liquor_venues[0])
```

```python
!pip install pandas folium
```

```python
import folium
import pandas as pd
processed_venues_data = []

for venue in liquor_venues:
    name = venue["properties"]["premises_name"]
    suburb = ''
    if suburb in venue["properties"].keys():
        suburb = venue["properties"]["suburb"]
    longitude = venue[ "geometry"]["coordinates"][0]
    latitude = venue[ "geometry"]["coordinates"][1]
    new_venue_data = {"name":name, "longitude":longitude, "latitude": latitude, "suburb": suburb}
    processed_venues_data.append(new_venue_data)


df = pd.DataFrame(processed_venues_data[:1000])
df = pd.DataFrame(df[df['suburb'] != 'MELBOURNE'])

m = folium.Map(location=[df['latitude'].mean(), df['longitude'].mean()], zoom_start=10)


for index, row in df.iterrows():
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=row['name'],
        icon=folium.Icon(icon='info-sign')
    ).add_to(m)

m
```

## Visualise the car crashes

```python
from folium.plugins import HeatMap


df = pd.read_csv("../elastic/data/Road_Crashes_for_five_Years_Victoria.csv")
print(df.head())

df = df[df['LGA_NAME_ALL'] != 'MELBOURNE']


locations = df[['LATITUDE', 'LONGITUDE']]

# Drop any rows with missing values in these columns
locations = locations.dropna()

# Convert to a list of lists
location_data = locations.values.tolist()

# Initialize the map centered around the average location
m = folium.Map(location=[locations['LATITUDE'].mean(), locations['LONGITUDE'].mean()], zoom_start=10)

# Add the heat map layer
HeatMap(location_data).add_to(m)

for venue in liquor_venues:
    name = venue["properties"]["premises_name"]
    longitude = venue[ "geometry"]["coordinates"][0]
    latitude = venue[ "geometry"]["coordinates"][1]
    new_venue_data = {"name":name, "longitude":longitude, "latitude": latitude}
    processed_venues_data.append(new_venue_data)


df = pd.DataFrame(processed_venues_data[:150])


for index, row in df.iterrows():
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=row['name'],
        icon=folium.Icon(icon='info-sign')
    ).add_to(m)

m
```
